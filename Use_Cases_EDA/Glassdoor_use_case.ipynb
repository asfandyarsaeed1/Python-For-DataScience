### Now set up Job title as there are jr ,junior in the data and we need to remove those and make a separate columns

```python
### there are lot of other designation on Glass Door but I will focus on data related titles.
def play_title(title):
    if isinstance (title,str):
        if 'data scientist' in title.lower():
            return 'data scientist'
        elif 'data engineer' in title.lower():
            return 'data engineer'
        if 'analyst' in title.lower():
            return 'analyst'
        if 'machine learning' in title.lower():
            return 'mle'
        if 'manager' in title.lower():
            return 'manager'
        if 'director' in title.lower():
            return 'director'
        else:
            return 'na'
    return title
def senior(title):
    if 'jr' in title.lower() or 'jr.' in title.lower():
        return 'jr'
    elif '' in title.lower() or 'sr' in title.lower() or 'senior' in title.lower() or 'lead' in title.lower() or 'principal' in title.lower():
        return 'sr'
    elif 'sr.' in title.lower() or 'sr' in title.lower() or 'senior' in title.lower() or 'lead' in title.lower() or 'principal' in title.lower():
        return 'sr'
    else:
        return 'na'
    
```

```python
df['job_simp']=df['Job Title'].apply(play_title)
df['job_simp'].value_counts()
```

### Now lets see what is there in 'na' its ok if we dont need them but we should know about them.

```python
df['level_title']=df['Job Title'].apply(senior)
df['level_title'].unique()
df['level_title'].value_counts()
```
