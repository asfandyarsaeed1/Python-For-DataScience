## Lets convert Reviews to int type

```python
df['Reviews'].value_counts().head()

## changing dataype of reviews

df[df['Reviews'].str.contains("M")].head()
```

```python
def converts(reviews):
    if isinstance(reviews,str):
        if '3.0M' in reviews:
            return int(int(reviews.replace('3.0M','3'))*1000000)
        else:
            return int(reviews)
    return reviews
```

```python
df['Reviews']=df['Reviews'].apply(converts)
```
